```python
# Temel kütüphaneler
# KUTUPHANELERIN CALISMASI ICIN ONCEDEN TERMINALDE SU KODU CALISTIRMALISINIZ

#pip install opencv-python mediapipe numpy

import cv2 #Görüntü Işleme Kütüphanemiz
import mediapipe as mp #Gözü tespit eden kütüphane
import numpy as np #Koordinat sistemine döken kütüphane
from google.colab.patches import cv2_imshow #EGER HATA ALIYORSAN SIL OYLE CALISTIR.

kamera = cv2.VideoCapture(0) #Bilgisayara kamerayı açmasını söylüyoruz. Burada ki 0 değeri varsayılan kamerayı kullanması gerektiğini söylüyor.

mp_yuz_agi = mp.solutions.face_mesh

yuz_agi = mp_yuz_agi.FaceMesh(
    max_num_faces=1,
    refine_landmarks=True,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)

while True: #Döngüye giriyoruz
    basarilimi, kare = kamera.read() #Kameradaki o an ki görüntüyü(kareyi) okuyoruz.
    # basarilimi değişkeni görüntü alındıysa true alınamadıysa false olur."kare" görüntünün kendisi.

    if not basarilimi: #Önce olumsuz durumu alıyoruz verimlilik için.
        print("Kamera Görüntüsü Alinamadi.")
        break #Kameradan görüntü alamıyorsak döngüyü bitirip çıkıyoruz.

    rgb_kare = cv2.cvtColor(kare, cv2.COLOR_BGR2RGB)

    sonuclar = yuz_agi.process(rgb_kare)

    if sonuclar.multi_face_landmarks:

        h, w, _ = kare.shape

        for yuz_noktalari in sonuclar.multi_face_landmarks:

            sol_goz_bebegi = yuz_noktalari.landmark[468]

            sol_x = int(sol_goz_bebegi.x * w)
            sol_y = int(sol_goz_bebegi.y * h)

            cv2.circle(kare, (sol_x, sol_y), 3, (0, 255, 0), -1)

            sag_goz_bebegi = yuz_noktalari.landmark[473]

            sag_x = int(sag_goz_bebegi.x * w)
            sag_y = int(sag_goz_bebegi.y * h)

            cv2.circle(kare,(sag_x, sag_y), 3, (0, 255,0), -1)


    cv2.imshow("Proje Kamerasi", kare)#Açılan kareyi küçük bir pencereden gösteriyoruz.
                                      #Tırnak içindeki kısım pencerenin ismi.


    if cv2.waitKey(1) & 0xFF == 27: #27 numarası klavyede esc tuşuna denk gelmekte. Esc ye bastıgında cıkıyor. waitkey(1) kapatmadan önce "1" milisaniye bekle demek. "0" DENEME
        break #döngüyü bitir.

kamera.release() #kamerayı sal
cv2.destroyAllWindows() #pencereyi kapat.
```
